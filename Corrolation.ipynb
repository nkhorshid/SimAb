```python
#Load data:
name_s = 'input/N_Atom.txt'
solar = np.loadtxt(name_s,dtype= str)[0:, 5].astype(float)

doc = 'Output/run_sum'+str(name)+'.txt'
doc1 = 'Output/run_sum'+str(name1)+'.txt'

dat1 = np.loadtxt(doc, dtype = str, skiprows = 1)[0:,1:].astype(float)
dat2 = np.loadtxt(doc1, dtype = str, skiprows = 1)[0:,1:].astype(float)

#test = np.loadtxt(doc, dtype = str, skiprows = 1)[0:,0]
#header = np.loadtxt(doc, dtype = str)[0,1:]
```

```python
#%matplotlib qt
c = dsk.r_arr[4:]/var.au
t = dsk.T_arr[4:]

g0 =np.where((dat[:,2]<=c[0]))
g1 =np.where((dat[:,2]<=c[1])&(dat[:,2]>c[0]))
g2 =np.where((dat[:,2]<=c[2])&(dat[:,2]>c[1]))
g3 =np.where((dat[:,2]<=c[3])&(dat[:,2]>c[2]))
g4 =np.where((dat[:,2]>c[3]))

datg0 = np.copy(dat[g0])
datg1 = np.copy(dat[g1])
datg2 = np.copy(dat[g2])
datg3 = np.copy(dat[g3])
datg4 = np.copy(dat[g4])
```

## Calculate corrolation for individual data:

```python
#['M_c =0' 'M_fin =1' 'a_c =2' 'a_fin =3' 'dstg_r =4' 'pls_r =5' 'temperature =6' 'metalicity =7'
#'H =8' 'He =9' 'Li =10' 'C =11' 'N =12' 'O =13' 'Na =14' 'Mg =15' 'Al =16' 'Si =17' 'S =18' 'K =19' 'Fe =20'
#'alpha =21']
```

```python
r  = (dat[:,11]*solar[3]/(dat[:,13]*solar[5]),dat[:,7])
r0 = (datg0[:,11]*solar[3]/(datg0[:,13]*solar[5]),datg0[:,7])
r1 = (datg1[:,11]*solar[3]/(datg1[:,13]*solar[5]),datg1[:,7])
r2 = (datg2[:,11]*solar[3]/(datg2[:,13]*solar[5]),datg2[:,7])
r3 = (datg3[:,11]*solar[3]/(datg3[:,13]*solar[5]),datg3[:,7])
r4 = (datg4[:,11]*solar[3]/(datg4[:,13]*solar[5]),datg4[:,7])
print (np.shape(r[0][:]))
```

```python
i = 5
j = 1

corr,e = spearmanr(dat[:,i],r[j])
corr0,e0 = spearmanr(datg0[:,i],r0[j])
corr1,e1 = spearmanr(datg1[:,i],r1[j])
corr2,e2 = spearmanr(datg2[:,i],r2[j])
corr3,e3 = spearmanr(datg3[:,i],r3[j])
corr4,e4 = spearmanr(datg4[:,i],r4[j])
print (corr,'\t',e,'\n',corr0,'\t',e0,'\n',corr1,'\t',e1,'\n',corr2,'\t',e2,'\n',corr3,'\t',e3,'\n',corr4,'\t',e4)
```
